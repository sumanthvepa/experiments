```python
import plotly.express as px
import pandas as pd

# Sample data
df = pd.DataFrame({
    "x": range(100),
    "y": [2*i*i + 4*i + 50 for i in range(100)]
})

fig = px.line(df, x="x", y="y", title="Interactive Quadratic Plot")
fig.show()
```

Here's a scatter plot:

```python
import pandas as pd
import plotly.express as px

# Example dataset
data = {
    "x": [1, 2, 3, 4, 5],
    "y": [10, 25, 18, 23, 29],
    "category": ["A", "C", "A", "B", "A"]
}
df = pd.DataFrame(data)

# Create scatter plot
fig = px.scatter(
    df, 
    x="x", 
    y="y", 
    color="category",   # optional: color points by category
    size="y",           # optional: vary point size
    hover_name="category"  # optional: label on hover
)

# Show the plot
fig.show()
```
